# Columns Description:
Row ID: Unique identifier for each row/record.

Order ID: Unique identifier for each order.

Order Date: The date when the order was placed.

Ship Date: The date when the order was shipped.

Ship Mode: The shipping method used (e.g., Second Class, Standard Class).

Customer ID: Unique identifier for each customer.

Customer Name: The name of the customer.

Segment: The market segment of the customer (Consumer, Corporate, Home Office).

Country: The country where the order was placed (primarily United States).

City: The city where the order was delivered.

State: The state where the order was delivered.

Postal Code: Postal code of the delivery address.

Region: The region in which the city is located (e.g., West, South).

Product ID: Unique identifier for each product.

Category: The main category of the product (Furniture, Office Supplies, Technology).

Sub-Category: A more specific classification within each category (e.g., Chairs, Labels).

Product Name: The name of the product.

Sales: The total sales amount for that particular product in the order.

## PERFORMING EXPLORATORY DATA ANALYSIS

```python
# Plotting the Sales vs. Region
plt.figure(figsize=(8, 6))
sns.barplot(x='Region', y='Sales', data=df, palette='viridis')

plt.title('Total Sales by Region', fontsize=18)
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Observation:

The West and East regions generally show higher total sales compared to the South and Central regions.

```python
# Plotting the Sales vs. State
plt.figure(figsize=(12, 8))
sns.barplot(x='Sales', y='State', data=df, palette='coolwarm')
plt.title('Total Sales by State', fontsize=18)
plt.xlabel('Total Sales')
plt.ylabel('State')
plt.tight_layout()
plt.show()
```

Observation:

Wyoming, Vermont and Nevada have the highest total sales, indicating strong market presence in these states.

```python
# Create a pivot table to sum sales by Category and Sub-Category
category_sales_pivot = df.pivot_table(values='Sales', index='Sub-Category', columns='Category', aggfunc='sum')

category_sales_pivot.plot(kind='barh', stacked=True, figsize=(14, 8), colormap='Set2')
plt.title('Total Sales by Category and Sub-Category', fontsize=18)
plt.xlabel('Total Sales')
plt.ylabel('Sub-Category')
plt.legend(title='Category')
plt.tight_layout()
plt.show()
```

Observations:

Sub-categories like Phones, storage and chairs under Technology show the highest sales contributions.

 Sub-categories like fasteners, labels have lower sales despite bulk orders.

```python
# Calculate total sales and profit by category
category_sales_profit = df.groupby('Category')[['Sales', 'Profit']].sum()

# Plot pie charts
fig, axes = plt.subplots(1, 2, figsize=(14, 7))

axes[0].pie(category_sales_profit['Sales'], labels=category_sales_profit.index, autopct='%1.1f%%', startangle=90)
axes[0].set_title('Sales by Category', fontsize=18)
axes[1].pie(category_sales_profit['Profit'], labels=category_sales_profit.index, autopct='%1.1f%%', startangle=90)
axes[1].set_title('Profit by Category', fontsize=18)
plt.tight_layout()
plt.show()
```

```python
ship_mode_profit = df.groupby(['Ship Mode', 'Year'])['Profit'].sum().unstack().fillna(0)
ship_mode_profit.plot(kind='bar', stacked=True, figsize=(14, 8), colormap='Set2')
plt.title('Profit Distribution by Ship Mode', fontsize=18)
plt.xlabel('Ship Mode')
plt.ylabel('Total Profit')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Observation:

The 'Standard Class' shipping mode likely shows the highest total profit due to its widespread use and cost-effectiveness. 

'Same Day' shipping, while faster, may yield lower profits because of higher operational costs.

```python
# Scatter plot to compare Discount vs Quantity
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Discount', y='Quantity', data=df, hue='Category', palette='Set2')
plt.title('Discount vs Quantity by Category', fontsize=18)
plt.xlabel('Discount')
plt.ylabel('Quantity')
plt.legend(title='Category')
plt.tight_layout()
plt.show()
```

Observation:

Higher discounts generally correlate with increased quantities purchased, especially in certain categories like Office Supplies.

However, the relationship may not be consistent across all categories—Technology products might show less sensitivity to discounts, while Furniture could have scattered patterns, indicating varying customer responses.

```python
# Calculate total sales by Customer Segment
sales_by_segment = df.groupby('Segment')['Sales'].sum()
plt.figure(figsize=(8, 4))
plt.pie(sales_by_segment, labels=sales_by_segment.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
centre_circle = plt.Circle((0,0), 0.50, fc='white')
plt.gca().add_artist(centre_circle)
plt.title('Sales Distribution by Customer Segment', fontsize=18)
plt.tight_layout()
plt.show()
```

Observation:

The Consumer segment likely contributes the largest share of total sales, followed by Corporate and Home Office segments. 

```python
# Ensure Order Date is in datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Extract Year and Month from Order Date
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month_name()

# Calculate total sales by Month and Year
monthly_sales = df.groupby(['Year', 'Month'])['Sales'].sum().reset_index()

# Sort by Year and Month for better visualization
monthly_sales['Month'] = pd.to_datetime(monthly_sales['Month'], format='%B').dt.month
monthly_sales = monthly_sales.sort_values(by=['Year', 'Month'])
         
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Month', y='Sales', hue='Year', data=monthly_sales, marker='o', palette='Set2')
plt.title('Total Sales by Month and Year', fontsize=18)
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout() 
plt.show()  
```

Observation:

Sales typically show a noticeable increase towards the end of the year, likely due to holiday shopping and year-end promotions.

Mid-year months may have moderate sales, while the early months often reflect slower activity.
